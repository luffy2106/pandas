5.6 [Turing interview] What is the neighborhood in which superhosts have the biggest price difference with respect to non superhosts?(Take only 300 data point to test to reduce time)

Paraphase the questions :

Among the neighborhood, which one has the biggest difference between superhosts and non-superhosts ?

Solution:
- Loop all the neighborhood, calcuate the substract of the price between superhosts and non-superhosts
- Choose the substraction which is biggest, this the the neighborhood we need 

Note:

Since the data is big, you should do this test in 10% of original dataset only




```python
listings_subset = listings.sample(frac=0.1, random_state=1)
```

```python
listings_subset['price'].head(3)
```

```python
listings_subset.columns
```

```python
listings_subset['host_is_superhost'].value_counts()
```

```python
list_neighbor = list(listings_subset['neighbourhood_cleansed'])
```

```python
listings_subset['neighbourhood_cleansed'] 
```

```python
# This code take 21m45s to execute
neighbor_most_price_difference = None
sum_difference_price = 0
for neighbor in list_neighbor:
    sum_price_neighbor_superhost = listings[(listings['host_is_superhost'] == "t") & (listings['neighbourhood_cleansed'] == neighbor)]['price'].sum()
    sum_price_neighbor_non_superhost = listings[(listings['host_is_superhost'] == "f") & (listings['neighbourhood_cleansed'] == neighbor)]['price'].sum()
    sum_difference_price_neighbor = sum_price_neighbor_superhost - sum_price_neighbor_non_superhost
    if sum_difference_price > sum_difference_price_neighbor:
        sum_difference_price = sum_difference_price_neighbor
        neighbor_most_price_difference = neighbor
```

```python
print("the neighborhood in which superhosts have the biggest price difference with respect to non superhosts is {}".format(neighbor_most_price_difference))
```

5.7 [Turing interview]. Which of the review scores has the highest correlation to price?

Paraphase the questions :

Among the type of reviews, which one has the highest correlation to price?

Solution:
- Calculate correlation between each column represent for one type of review and price 
- Choose the correlation which is the highest




```python
listings.columns
```

```python
list_reviews = ['review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value']
max_correlation_review = -1
highest_correlation_review = ""
for review in list_reviews:
    correlation_review = listings[review].corr(listings['price'])
    if correlation_review > max_correlation_review:
        max_correlation_review = correlation_review
        highest_correlation_review = review 
```

5.9[Turing interview] What is the median price premium given to entire homes / entire apartments with respect to other listings of the same neighborhood? Report the average across all neighborhoods.

Note:
- Price premium : the additional amount of money that a customer is willing to pay for a product or service compared to the base or standard price. Ex : Free version 10u, premium version 15u => price premium = 15-10=5u

Solution :
1. First we need to group listings of the same neighborhood
2. For listings in the same group, we need to calculate the following info :
- Median price of all listing with all room_type 
- Median price of room_type which is  "Entire home/apt"
- Find the price premium by subtracting the median price of other listings from the median price of "entire homes/apartments"(the ideas is they want to know that what is the premium price that users paid more when they choose room_type "entire homes/apartments" compare to other room_type)
3. Calculate the average price premium across all neighborhoods by taking the mean of the price premiums calculated in the previous step.

```python
listings['room_type'].value_counts()
```

```python
print(listings.head(4).to_markdown())
```

```python
# Grouping by Neighborhood
neighborhood_groups =  listings.groupby("neighbourhood_cleansed")
```

```python
type(neighborhood_groups)
```

```python
# # Uncomment if you want to see each group  
# for name, group in neighborhood_groups:
#     print(f"Neighborhood: {name}")
#     display(group.head(3))
```

```python
# Calculating Price Premium
price_premiums = []
for name, group in neighborhood_groups:
    entire_homes_price = group[group['room_type']== 'Entire home/apt']['price'].median()
    all_listings_price = group['price'].median()
    price_premium = entire_homes_price - all_listings_price
    price_premiums.append(price_premium)
```

```python
# Average Price Premium Calculation
average_price_premium = sum(price_premiums) / len(price_premiums)

# Report the Result
print(f"The average price premium for entire homes/apartments across different neighborhoods is: ${average_price_premium:.2f}")
```

# 10. Handling outliers

10.1 Find the outliers which is to cheap(less than 25% of price) and too expensive(more than 75% of the price) of dataframe listings

```python
Q1 = listings['price'].quantile(0.25)
Q3 = listings['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = listings[(listings["price"] < lower_bound) | (listings['price'] > upper_bound)]
```

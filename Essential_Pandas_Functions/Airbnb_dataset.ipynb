4.1 Convert column "price" to float in dataframe "listings"

```python
listings['price']
```

```python
def convert_str_to_float(x):
    x = x.replace('$','')
    x = x.replace(',','')
    return float(x)

def convert_str_to_float_by_regex(x):
    pattern = r'[$,]'
    x = re.sub(pattern, '', x)
    return float(x)
```

```python
listings['price'] = listings['price'].apply(convert_str_to_float)
```

2.5 What is the neighborhood in which superhosts have the biggest price difference with respect to non superhosts?

Paraphase the questions :

Among the neighborhood, which one has the biggest difference between superhosts and non-superhosts ?

Solution:
- Loop all the neighborhood, calcuate the substract of the price between superhosts and non-superhosts
- Choose the substraction which is biggest, this the the neighborhood we need 


```python
listings['price'].head(3)
```

```python
listings.columns
```

```python
listings['host_is_superhost'].value_counts()
```

```python
list_neighbor = list(listings['neighbourhood_cleansed'])
```

```python
listings['neighbourhood_cleansed'] 
```

```python
# This code take 21m45s to execute
neighbor_most_price_difference = None
sum_difference_price = 0
for neighbor in list_neighbor:
    sum_price_neighbor_superhost = listings[(listings['host_is_superhost'] == "t") & (listings['neighbourhood_cleansed'] == neighbor)]['price'].sum()
    sum_price_neighbor_non_superhost = listings[(listings['host_is_superhost'] == "f") & (listings['neighbourhood_cleansed'] == neighbor)]['price'].sum()
    sum_difference_price_neighbor = sum_price_neighbor_superhost - sum_price_neighbor_non_superhost
    if sum_difference_price > sum_difference_price_neighbor:
        sum_difference_price = sum_difference_price_neighbor
        neighbor_most_price_difference = neighbor
```

```python
print("the neighborhood in which superhosts have the biggest price difference with respect to non superhosts is {}".format(neighbor_most_price_difference))
```
